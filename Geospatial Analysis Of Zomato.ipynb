Nominatim is a tool to search OpenStreetMap data by address or location

```python
from geopy.geocoders import Nominatim
```

```python
geo_loc=Nominatim(user_agent='app',timeout=None)
```

Extracting Coordinates from Restaurant's Location

```python
lat=[]
long=[]
```

```python
for loc in rest['location']:
    location=geo_loc.geocode(loc)
    if location is None:
        lat.append(np.nan)
        long.append(np.nan)
    else:
        lat.append(location.latitude)
        long.append(location.longitude)
```

```python
rest['latitude']=lat
rest['longitude']=long
```

```python
def rating(x):
    if pd.isnull(x):
        return np.nan
    elif x=='NEW':
        return 0
    elif x=='-':
        return 0
    else:
        return x.split('/')[0]
```

```python
df['rating']=df['rate'].apply(rating)
```

```python
df['rating'].unique()
```

```python
df['rating'].isnull().sum()/len(df['rating']) *100
```

About 15% of values are missing so we cannot drop it as it has significant impact on the operations.

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
mean_ratings = df.groupby('name')['rating'].transform('mean')
```

```python
mean_ratings
```

```python
df['rating'].fillna(mean_ratings, inplace=True)
```

```python
filter=df.groupby('name')['rating'].mean()
```

```python
filter
```

```python
s=filter.sort_values(ascending=False)
```

```python
s
```

```python
len(df['name'].unique())
```

```python
df_new=pd.DataFrame()
```

```python
df_new['name']=df['name'].unique()
```

```python
v=df['name'].value_counts()
```

```python
df_new['no_of_orders']=df['name'].apply(lambda x: v[x])
```

```python
df_new['avg_rating']=df['name'].apply(lambda x: s[x])
```
